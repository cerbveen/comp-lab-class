```python
# this notebook is for Week3
# import libraries
import numpy as np              # numpy| data processing
import matplotlib.pyplot as plt # matplotlib| plotting
```

```python
# load data
DATA = np.loadtxt('./Data/potential.xvg',skiprows=24)
# numpy.loadtxt(fname,skiprows=)
```

```python
# figure 1
plt.figure(1)                           # create figure 1
# title & labels
plt.title('Potential vs. Time')         # title     
plt.xlabel('Time (ns)')                 # x-axis label
plt.ylabel('Potential (kJ/mol)')        # y-axis label
# plot, change x to ns, colorblind friendly, linestyle, linewidth
plt.plot(DATA[:,0]/1000,DATA[:,1],color='#0072B2',linestyle='-',linewidth=1)
# adjust axis range
# plt.xlim(0,50)                          # x-axis range
# change y-axis to scientific notation
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
# save figureï¼Œ overwrite protection
```

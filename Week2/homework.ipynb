```python
# figure 7
fig, ax1 = plt.subplots()                           # create figure 7
# title & labels
plt.suptitle('Temperature & Kinetic Engery vs. Time')       # title
# plot, colorblind friendly, linestyle, linewidth
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature (K)', color='#0072B2')
ax1.plot(DATA[:,0]/1000,DATA[:,4],color='#0072B2',linestyle='solid',linewidth=1)
ax1.tick_params(axis='y', labelcolor='#0072B2')
ax1.ticklabel_format(style='plain', axis='y', scilimits=(0,0), useMathText=True)
# second axis
ax2 = ax1.twinx()
ax2.set_ylabel('Kinetic Energy (kJ/mol)', color='#D55E00')
ax2.plot(DATA[:,0]/1000,DATA[:,2],color='#D55E00',linestyle='dashed',linewidth=1)
ax2.tick_params(axis='y', labelcolor='#D55E00')
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
# legend
fig.legend(['Temperature','Kinetic Energy'],loc='upper right',bbox_to_anchor=(1,1),bbox_transform=ax1.transAxes,frameon=False)
# adjust axis range
ax1.set_xlim(0,50)                          # x-axis range
# save figure， overwrite protection
fname = 'Part 4-Fig.7 Temperature&Kinetic_Energy_vs_Time.pdf'
if os.path.exists('./Homework/Plots/'+fname):
    print('File exists, overwrite?')
    print('1: yes, 0: no')
    overwrite = int(input())
    if overwrite == 1:
        plt.savefig('./Homework/Plots/'+fname)
        print('File saved')
        plt.show()                              
    else:
        print('File not saved')
else: 
    plt.savefig('./Homework/Plots/'+fname)
    print('File saved')
    plt.show()
```

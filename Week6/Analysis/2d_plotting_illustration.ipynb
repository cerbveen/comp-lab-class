This notebook is to explain how to analysis 2d-metad in lab 6


```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load data
setting='11.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_1.05_2d_metad_10ns'
path='../Data/'+setting+'/basin'+'A'+'_2d_metad.fes.dat'
data=np.loadtxt(path)
```

```python
# shape of data
print('shape of data:',data.shape)
```

The headings of data file
#! FIELDS phi psi file.free der_phi der_psi
#! SET min_phi -pi
#! SET max_phi pi
#! SET nbins_phi  [33]
#! SET periodic_phi true
#! SET min_psi -pi
#! SET max_psi pi
#! SET nbins_psi  [17]
#! SET periodic_psi true
------------------
In this file, there are [33] different values for phi, [17] different values for psi, [33*17=561] different values for free energy (file.free)

```python
# read data without reshaping
phi=data[:,0]
psi=data[:,1]
fes=data[:,2]
# print the first 10 rows
print('phi:',phi[:10])
print('psi:',psi[:10])
print('fes:',fes[:10])
# print the shape of data
print('shape of phi:',phi.shape)
print('shape of psi:',psi.shape)
print('shape of fes:',fes.shape)
```

```python
# reshape data
# phi_new is the first 33 rows of phi
phi_new=phi[:33]
# psi_new is every 33 rows of psi
psi_new=psi[::33]
# fes_new needs to be reshaped
fes_new=fes.reshape(17,33)

# print the first 10 rows
print('phi_new:',phi_new[:10])
print('psi_new:',psi_new[:10])

# print the shape of data
print('shape of phi_new:',phi_new.shape)
print('shape of psi_new:',psi_new.shape)
print('shape of fes_new:',fes_new.shape)
```

```python
# plot data wiht imshow
plt.imshow(fes_new,origin='lower',extent=[phi_new[0],phi_new[-1],psi_new[0],psi_new[-1]])
```

```python
# plot data with contourf
plt.contourf(phi_new,psi_new,fes_new,20)
```

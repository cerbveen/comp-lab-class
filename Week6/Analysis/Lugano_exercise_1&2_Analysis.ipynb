```python
# figure 5 - bias landscape & free energy landscape
fig,axs=plt.subplots(1,2,layout=None, figsize=(12,5))
# data processing, load time in ns
phi_1 = grid_1[:,0]
bias_1 = grid_1[:,1]
derivatives_phi_1 = grid_1[:,2]
biasfactor_1 = 10
free_energy_1 = bias_1 * (biasfactor_1/(1-biasfactor_1))
free_energy_derivatives_phi_1 = derivatives_phi_1 * (biasfactor_1/(1-biasfactor_1))

# set colorblind-friendly palatte
colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB']

# plot minus bias potential over phi, colorblind friendly, line width 2
axs[0].plot(phi_1, -bias_1, c=colors[0], lw=2, label='minus bias potential ''-V')
# set axis range
axs[0].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[0].set_xlabel(r'$\phi$ (rad)', fontsize=12)
axs[0].set_ylabel('Minus Bias Potential (kJ/mol)', fontsize=12)
# title
axs[0].set_title("Minus bias potential landscape", fontsize=12)
# plot derivatives of minus bias potential over phi, colorblind friendly, line width 1, dashed, right y axis
ax0_2=axs[0].twinx()
ax0_2.plot(phi_1, -derivatives_phi_1, c=colors[1], lw=1, ls='--', label='minus derivative 'r'$-\frac{\partial V}{\partial \phi}$')
# labels, font size 12
ax0_2.set_ylabel('Derivative (kJ/mol'r'$\bullet$''rad))', fontsize=12)
# combine legend for twinx
axs[0].legend(loc='upper left', bbox_to_anchor=(0,1), bbox_transform=axs[0].transAxes, fontsize=12, frameon=False)
ax0_2.legend(loc='upper left', bbox_to_anchor=(-0.003,0.95), bbox_transform=axs[0].transAxes, fontsize=12, frameon=False)

# plot free energy over phi, colorblind friendly, line width 2
axs[1].plot(phi_1, free_energy_1, c=colors[0], lw=2, label='free energy 'r'$F$')
# set axis range
axs[1].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[1].set_xlabel(r'$\phi$ (rad)', fontsize=12)
axs[1].set_ylabel('Free Energy (kJ/mol)', fontsize=12)
# title
axs[1].set_title("Free energy landscape", fontsize=12)
# plot derivatives of free energy over phi, colorblind friendly, line width 1, dashed, right y axis
ax1_2=axs[1].twinx()
ax1_2.plot(phi_1, free_energy_derivatives_phi_1, c=colors[1], lw=1, ls='--', label='derivative 'r'$\frac{\partial F}{\partial \phi}$')
# labels, font size 12
ax1_2.set_ylabel('Derivative (kJ/mol'r'$\bullet$''rad))', fontsize=12)
# combine legend for twinx
axs[1].legend(loc='upper left', bbox_to_anchor=(0,1.01), bbox_transform=axs[1].transAxes, fontsize=12, frameon=False)
ax1_2.legend(loc='upper left', bbox_to_anchor=(-0.003,0.94), bbox_transform=axs[1].transAxes, fontsize=12, frameon=False)

# title
plt.text(x=0.5, y=0.95, s="Bias potential & free energy landscape", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "ADP, 10 ns metad, biasfactor10_pace500_height12_sigma0.55", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.4)

# save figure as pdf in ../Figures
fname = '../Figures/fig5.[metad_10ns_topolA]Energy_landscape.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
# load data

# path setting & strides setting
# strides are 1, 3, 10, 20, 50, 100 
settings_sigma_comparison=['1.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_metad_10ns',
'2.adp_basinA_biasfactor10_pace500_height1.2_sigma0.25_metad_10ns',
'3.adp_basinA_biasfactor10_pace500_height1.2_sigma0.85_metad_10ns']
sigmas=['0.55', '0.25', '0.85']
settings_height_comparison=['1.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_metad_10ns',
'4.adp_basinA_biasfactor10_pace500_height0.6_sigma0.55_metad_10ns',
'5.adp_basinA_biasfactor10_pace500_height1.8_sigma0.55_metad_10ns']
heights=['1.2', '0.6', '1.8']
strides=['1','3','10','20','50','100']

# create matrix to store path
fes_final_path_sigma_comparison = np.empty((len(settings_sigma_comparison)), dtype=object)
fes_final_path_height_comparison = np.empty((len(settings_height_comparison)), dtype=object)
fes_stride_path_sigma_comparison = np.empty((len(settings_sigma_comparison), len(strides)), dtype=object)
fes_stride_path_height_comparison = np.empty((len(settings_height_comparison), len(strides)), dtype=object)

# final fes path
for i in range(len(settings_sigma_comparison)):
    fes_final_path_sigma_comparison[i] = '../Data/'+settings_sigma_comparison[i]+'/'+settings_sigma_comparison[i]+'.fes.dat'
    i += 1
#print (fes_final_path_sigma_comparison)
for i in range(len(settings_height_comparison)):
    fes_final_path_height_comparison[i] = '../Data/'+settings_height_comparison[i]+'/'+settings_height_comparison[i]+'.fes.dat'
    i += 1
#print (fes_final_path_height_comparison)

# stride fes path
for i in range(len(settings_sigma_comparison)):
    for j in range(len(strides)):
        fes_stride_path_sigma_comparison[i,j] = '../Data/'+settings_sigma_comparison[i]+'/stride_hills/fes.dat'+strides[j]+'.dat'
        j += 1
    i += 1
#print (fes_stride_path_sigma_comparison)
for i in range(len(settings_height_comparison)):
    for j in range(len(strides)):
        fes_stride_path_height_comparison[i,j] = '../Data/'+settings_height_comparison[i]+'/stride_hills/fes.dat'+strides[j]+'.dat'
        j += 1
    i += 1
#print (fes_stride_path_height_comparison)

# create matrix to store data
fes_final_sigma_comparison = np.empty((len(settings_sigma_comparison)), dtype=object)
fes_final_height_comparison = np.empty((len(settings_height_comparison)), dtype=object)
fes_stride_sigma_comparison = np.empty((len(settings_sigma_comparison), len(strides)), dtype=object)
fes_stride_height_comparison = np.empty((len(settings_height_comparison), len(strides)), dtype=object)

# load final fes data
for i in range(len(settings_sigma_comparison)):
    fes_final_sigma_comparison[i] = np.loadtxt(fes_final_path_sigma_comparison[i])
    i += 1
for i in range(len(settings_height_comparison)):
    fes_final_height_comparison[i] = np.loadtxt(fes_final_path_height_comparison[i])
    i += 1

# load stride fes data
for i in range(len(settings_sigma_comparison)):
    for j in range(len(strides)):
        fes_stride_sigma_comparison[i,j] = np.loadtxt(fes_stride_path_sigma_comparison[i,j])
        j += 1
    i += 1
for i in range(len(settings_height_comparison)):
    for j in range(len(strides)):
        fes_stride_height_comparison[i,j] = np.loadtxt(fes_stride_path_height_comparison[i,j])
        j += 1
    i += 1

# check data
print(fes_final_sigma_comparison[0].shape)
print(fes_final_height_comparison[0].shape)
print(fes_stride_sigma_comparison[0,0].shape)
print(fes_stride_height_comparison[0,0].shape)
```

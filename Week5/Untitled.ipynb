```python
# this notebook is for Week3-Assignment-remove water from trajectory file
# author: Rui Zeng, N15471955, rz2522@nyu.edu
# import libraries
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
trr_file = './Inputs/T300/adp_exchange4temps.trr'
gro_file = './Inputs/T300/adp_exchange4temps.gro'
# load the trajectory
traj = md.load(trr_file, top=gro_file)
#traj = md.load('\Inputs\adp.trr', top='\Inputs\adp.gro')
# compute phi and psi angles
phi= md.compute_phi(traj)
psi= md.compute_psi(traj)
```

```python
phi[1].max()
```

```python
# plt the phi and psi angles vs time
plt.plot(traj.time,phi[1])
plt.show()
plt.plot(psi[1])
plt.show()
```

```python
# compute 2D histogram
#hist, xedges, yedges = np.histogram2d(phi, psi, bins=50)
hist, xedges, yedges = np.histogram2d(phi[1].flatten(), psi[1].flatten(), bins=100)
hist = -np.log(hist.T)
hist = hist-np.min(hist)
# plot the 2D histogram
plt.imshow(hist, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.show()
```

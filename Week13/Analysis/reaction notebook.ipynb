```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# path to data
path_standard= "../Outputs/monitor/MONITOR-COLVAR.metadynLog"
path_metadyn= "../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog"
path_metadyn_freqhills= "../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog"

# read data
data_standard = np.loadtxt(path_standard)
data_metadyn = np.loadtxt(path_metadyn)
data_metadyn_freqhills = np.loadtxt(path_metadyn_freqhills)
```

```python
# figure 6 - reaction coordinate
plt.figure(6)

# set colorblind-friendly palatte 
colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB']
line_width = [2,1.5,1]

# plot standard
plt.plot(data_standard[190:,0]/1000, data_standard[190:,1], color=colors[0], label='standard QM/MM', linewidth=line_width[0])
# plot metadyn
plt.plot(data_metadyn[:,0]/1000, data_metadyn[:,1], color=colors[1], label='metad: freq 1/100', linewidth=line_width[1])
# plot metadyn_freqhills
plt.plot(data_metadyn_freqhills[:,0]/1000, data_metadyn_freqhills[:,1], color=colors[2], label='metad: freq 1/20', linewidth=line_width[2])

# add horizontal line at y=0
plt.axhline(y=0, color='#BBBBBB', linestyle='--', linewidth=1)

# set labels
plt.xlabel('Time (ps)', fontsize=12)
plt.ylabel('Reaction coordinate $d_1 - d_2$ (nm)', fontsize=12)
# set x and y limits
plt.xlim(0, 30)
plt.ylim(-20, 11)

# set legend
plt.legend(loc='upper right', fontsize=12, frameon=False)

# save figure as pdf in ../Figures
fname = '../Figures/fig6.reaction coordinate.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')

```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# path to data
energy_path_NPT = "../Outputs/NPT-1.ener"
energy_path_NVT = "../Outputs/NVT-1.ener"
energy_path_standard = "../Outputs/monitor/MONITOR-1.ener"
energy_path_metadyn = "../Outputs/metad1_initmonitor/METADYN-1.ener"
energy_path_metadyn_freqhills = "../Outputs/metad1_initmonitor_freqhills/METADYN-1.ener"

# read data
energy_NPT = np.loadtxt(energy_path_NPT, skiprows=1)
energy_NVT = np.loadtxt(energy_path_NVT, skiprows=1)
energy_standard = np.loadtxt(energy_path_standard, skiprows=1)
energy_metadyn = np.loadtxt(energy_path_metadyn, skiprows=1)
energy_metadyn_freqhills = np.loadtxt(energy_path_metadyn_freqhills, skiprows=1)


```

```python
# figure 1 - potential energy vs time for NPT
plt.figure(1)
plt.plot(energy_NPT[:,1]/1000, energy_NPT[:,4], label="NPT")
plt.xlabel("Time (ps)",fontsize=12)
plt.ylabel("Potential Energy (a.u.)",fontsize=12)
# title
plt.title("Potential Energy vs Time for NPT")

# save figure as pdf in ../Figures
fname = '../Figures/fig1.potential_energy_NPT.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
# figure 2 - potential energy vs time for NVT
plt.figure(2)
plt.plot(energy_NVT[:,1]/1000, energy_NVT[:,4], label="NVT")
plt.xlabel("Time (ps)",fontsize=12)
plt.ylabel("Potential Energy (a.u.)",fontsize=12)
# title
plt.title("Potential Energy vs Time for NVT")

# save figure as pdf in ../Figures
fname = '../Figures/fig2.potential_energy_NVT.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
# figure 3 - potential energy vs time for standard QM/MM
plt.figure(3)
plt.plot(energy_standard[:,1]/1000, energy_standard[:,4], label="standard")
plt.xlabel("Time (ps)",fontsize=12)
plt.ylabel("Potential Energy (a.u.)",fontsize=12)
# title
plt.title("Potential Energy vs Time for standard QM/MM")

# save figure as pdf in ../Figures
fname = '../Figures/fig3.potential_energy_standard.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
# figure 4 - potential energy vs time for metadynamics
plt.figure(4)
plt.plot(energy_metadyn[:,1]/1000, energy_metadyn[:,4], label="metadynamics")
plt.xlabel("Time (ps)",fontsize=12)
plt.ylabel("Potential Energy (a.u.)",fontsize=12)
# title
plt.title("Potential Energy vs Time for metadynamics")

# save figure as pdf in ../Figures
fname = '../Figures/fig4.potential_energy_metadyn.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```
